### Exercício 1 – Condicional Simples

```python
resposta = input("Você já montou sua árvore de Natal? sim ou nao: ")

if resposta.lower() == "sim":
    print("Que ótimo! Sua casa já está no clima natalino")
else:
    print("Ainda dá tempo! Monte sua árvore hoje e deixe o espírito do Natal entrar! ")
```

### Exercício 2 – Laço de Repetição

```python
num = int(input("Digite um número para a contagem regressiva: "))

for p in range(num, 0, -1):
    print(p)

print("O Natal está chegando!")
```

### Exercício 3 – Funções + Lista + Condicional

```python
def verificar_presentes(lista_presentes):

    quantidade = len(lista_presentes)

    if quantidade >= 3:
        print("O Papai Noel terá trabalho este ano!")
    else:
        print("Sua lista está muito curta!")

    return quantidade


presentes = ["Bicicleta", "Lego", "Jogo de tabuleiro"]
total = verificar_presentes(presentes)
print("Quantidade:", total)
```


### Exercício 4 – Funções + Dicionário + Laços

```python

enfeites = {
    "bolas": 12,
    "estrelas": 3,
    "pisca-pisca": 1,
    "laços": 8
}

def contar_enfeites(enfeites):
    total = 0

    for quantidade in enfeites.values():
        total = total + quantidade

    return total

print("Total de enfeites:", contar_enfeites(enfeites))
```

Exercício 5 – Função Completa (listas, dicionários, laço e condicional)

```python
dados = {}

qtd_pessoas = int(input("Quantas pessoas vão contribuir para a ceia? "))

for p in range(qtd_pessoas):
    nome = input("Nome da pessoa: ")

    qtd_pratos = int(input(f"Quantos pratos {nome} vai levar? "))

    pratos = []
    for i in range(qtd_pratos):
        prato = input(f"Digite o prato: ")
        pratos.append(prato)

    dados[nome] = pratos


def resumo_ceia(dados):
    total_pessoas = len(dados)
    total_pratos = 0

    for lista in dados.values():
        total_pratos = total_pratos + len(lista)

    print(f"\nTotal de pessoas contribuindo: {total_pessoas}")
    print(f"Total de pratos: {total_pratos}")

    if total_pratos > 10:
        print("Essa ceia vai ser farta!")


resumo_ceia(dados)
```
